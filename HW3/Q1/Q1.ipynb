# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import imageio as iio
import scipy.ndimage as ndimage
```

# Calculating SNR

```python
img = iio.imread('city_orig.jpg')
noisy = iio.imread('city_noise.jpg')
x, y = img.shape

def show_image(img, title='', save=False, name=''):
    '''
    a function for showing images
    '''
    plt.figure()
    plt.imshow(img, cmap='gray')
    plt.xticks([])
    plt.yticks([])
    plt.title(title)
    if save:
        plt.savefig(name)
    plt.show()
    
def SNR(img, noisy):
    '''
    a function for calculating SNR
    '''
    img = img.astype('float64').copy()
    noisy = noisy.astype('float64').copy()
    return 10*np.log10(np.sum(img**2)/np.sum((img-noisy)**2))
    
# SNR in upper left
img1 = img[0:x//2, 0:y//2].copy()
noisy1 = noisy[0:x//2, 0:y//2].copy()
print('SNR of upper left is: {:.4f}dB'.format(SNR(img1, noisy1)))

# SNR in lower left
img1 = img[x//2:x, 0:y//2].copy()
noisy1 = noisy[x//2:x, 0:y//2].copy()
print('SNR of lower left is: {:.4f}dB'.format(SNR(img1, noisy1)))

# SNR in lower right
img1 = img[x//2:x, y//2:y].copy()
noisy1 = noisy[x//2:x, y//2:y].copy()
print('SNR of lower right is: {:.4f}dB'.format(SNR(img1, noisy1)))
```

# Median Filter

```python
med = ndimage.median_filter(noisy, size=3)
show_image(med)

# SNR in upper left
img1 = img[0:x//2, 0:y//2].copy()
noisy1 = med[0:x//2, 0:y//2].copy()
print('SNR of upper left is: {:.4f}dB'.format(SNR(img1, noisy1)))

# SNR in lower left
img1 = img[x//2:x, 0:y//2].copy()
noisy1 = med[x//2:x, 0:y//2].copy()
print('SNR of lower left is: {:.4f}dB'.format(SNR(img1, noisy1)))

# SNR in lower right
img1 = img[x//2:x, y//2:y].copy()
noisy1 = med[x//2:x, y//2:y].copy()
print('SNR of lower right is: {:.4f}dB'.format(SNR(img1, noisy1)))
```

# Gaussian Filter

```python
gauss = ndimage.gaussian_filter(noisy, sigma=5)
show_image(gauss)

# SNR in upper left
img1 = img[0:x//2, 0:y//2].copy()
noisy1 = gauss[0:x//2, 0:y//2].copy()
print('SNR of upper left is: {:.4f}dB'.format(SNR(img1, noisy1)))

# SNR in lower left
img1 = img[x//2:x, 0:y//2].copy()
noisy1 = gauss[x//2:x, 0:y//2].copy()
print('SNR of lower left is: {:.4f}dB'.format(SNR(img1, noisy1)))

# SNR in lower right
img1 = img[x//2:x, y//2:y].copy()
noisy1 = gauss[x//2:x, y//2:y].copy()
print('SNR of lower right is: {:.4f}dB'.format(SNR(img1, noisy1)))
```

# Mean Filter

```python
avg = ndimage.median_filter(noisy, size=5)
show_image(avg)

# SNR in upper left
img1 = img[0:x//2, 0:y//2].copy()
noisy1 = avg[0:x//2, 0:y//2].copy()
print('SNR of upper left is: {:.4f}dB'.format(SNR(img1, noisy1)))

# SNR in lower left
img1 = img[x//2:x, 0:y//2].copy()
noisy1 = avg[x//2:x, 0:y//2].copy()
print('SNR of lower left is: {:.4f}dB'.format(SNR(img1, noisy1)))

# SNR in lower right
img1 = img[x//2:x, y//2:y].copy()
noisy1 = avg[x//2:x, y//2:y].copy()
print('SNR of lower right is: {:.4f}dB'.format(SNR(img1, noisy1)))
```
