## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import imageio as iio
import cv2
```

## Histogram Equalization

```python
img = iio.imread('q4.png')
he_img = img.copy()
for i in range(img.shape[2]):
    he_img[:,:,i] = cv2.equalizeHist(img[:,:,i])
plt.imshow(he_img)
plt.xticks([])
plt.yticks([])
iio.imwrite('q4_1.png', he_img)
```

## Adaptive Histogram Equalization

```python
img = iio.imread('q4.png')
ahe_img = img.copy()
for i in range(img.shape[2]):
    ahe = cv2.createCLAHE(clipLimit=1e3, tileGridSize=(3,3))
    ahe_img[:,:,i] = ahe.apply(img[:,:,i])
plt.imshow(ahe_img)
plt.xticks([])
plt.yticks([])
iio.imwrite('q4_2.png', ahe_img)
```

## Contrast Limited Adaptive Histogram Equalization

```python
img = iio.imread('q4.png')
clahe_img = img.copy()
for i in range(img.shape[2]):
    clahe = cv2.createCLAHE(clipLimit=2, tileGridSize=(3,3))
    clahe_img[:,:,i] = clahe.apply(img[:,:,i])
plt.imshow(clahe_img)
plt.xticks([])
plt.yticks([])
iio.imwrite('q4_3.png', clahe_img)
```
